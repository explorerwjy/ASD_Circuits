```python
%load_ext autoreload
%autoreload 2
from ASD_Circuits import *

graph = LoadConnectome2()
def TestProcedure(bias_rank_fil):
    struct_rank_fil = bias_rank_fil
    g, top_structs = LoadConnectome(struct_rank_fil, Bin=True)
    top_nodes = g.vs.select(label_in=top_structs)
    g2 = g.subgraph(top_nodes)
    
    edge_permute_stat = g2.ecount()
    g2_ecounts = EdgePermutation(g, top_nodes, edge_permute_stat, Npermute=100)
    PlotPermutationP(g2_ecounts, edge_permute_stat, "Edge Permutation Test")
    
    node_permute_stat = g2.ecount()
    node_null = NodePermutation(g, top_nodes, node_permute_stat, Npermute=100)
    PlotPermutationP(node_null, node_permute_stat, "Node Permutation Test")
```
