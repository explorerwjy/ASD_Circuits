```python
def ShowWeight(STR, DF, STR2REG, Source=True):
    if Source:
        XX = DF[STR].sort_values(ascending=False)
    else:
        XX = DF.loc[STR, :].sort_values(ascending=False)
    res = pd.DataFrame(data={'Region': [STR2REG[x] for x in XX.index.values], 'STR':XX.index.values, 'Weight':XX.values})
    return res
```
